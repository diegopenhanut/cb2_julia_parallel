```julia
@everywhere df = DataFrame(Origin = ["a","b","c","d","a","d"], 
Target = ["j", "f", "g", "h", "i", "j"])
```

```julia
@everywhere function randNdCol(mytable)
	p=randperm(length(mytable[2]))
	mytable[2]=mytable[2][p]
	return mytable
	end
```

```julia
M = [df for i=1:10]
pmap(randNdCol, M)
```
